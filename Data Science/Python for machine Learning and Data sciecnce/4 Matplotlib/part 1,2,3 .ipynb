```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
# functional method
plt.plot(x,y,)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# object oriented method 
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # item 0 in list is Left of the axis 1 is bottom of the axis
# item 3 is width and 4 is height of the axis

axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel(" Y label")
axes.set_title("Title")
```

```python
fig = plt.figure()


axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])


axes1.plot(x,y)
axes1.set_title("Larger one")
axes2.plot(y,x)
axes2.set_title("Smaller one ")
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
# now we are learning to plot subplots using Object oriented 
```

```python
fig,axes = plt.subplots(nrows = 1,ncols=2)

axes[0].plot(x,y) # plots in first graph
axes[0].set_title("First graph")
axes[1].plot(y,x) # plots in second graph 
axes[1].set_title("second graph")
```

## Figure size and DPI

```python
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# the same plot can be done using 
fig,ax = plt.subplots(figsize =(8,2),nrows = 2, ncols = 1)
ax[0].plot(x,y)

ax[1].plot(y,x)

plt.tight_layout()
```

```python
# create a legend , that means plot 2 or more plots in single graph 
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label = "x square ",)
ax.plot(x, x**3, label = "x Cubed")
ax.legend()
```

### Plot appearences (Adding colors to the plot)

```python
fig = plt.figure()

ax= fig.add_axes([0,0,1,1])
ax.plot(x,y,color = '#a1160a',linewidth = 3,linestyle = "--",marker = "o",markersize = 20,
       markerfacecolor = "yellow",markeredgewidth = '3',markeredgecolor= "blue")

#we can also use lw insted of linewidth
#linestyle helps to plot the different types of plots like dashed line 
#linestyle can also be written as ls
```

```python
fig,ax = plt.subplots(figsize =(12,6))


ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestyle options -, , -., :, steps
ax.plot(x, x+5 ,color = "green", lw = 2.5,linestyle = '--' )
ax.plot(x ,x+6 ,color = 'green', lw = 3,linestyle = "-.")
ax.plot(x, x+7 , color = 'green', lw = 3.5, linestyle = ":")


## possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
