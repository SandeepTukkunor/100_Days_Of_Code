```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#seaborn comes with built in data sets
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde = False,bins= 40)
```

```python
#joint plot
sns.jointplot(x ='total_bill' , y ='tip' , data = tips)
#kind = reg for regresson , kde will help with 2-d Density , hex allows to make hexagonal distribution 

```

```python
#paitplot , plots pair wise relations ie plot of all the different type of data in the data set 

sns.pairplot(tips,hue = 'sex', palette = "coolwarm")
#hue helps to specify the difference like catogorical data 
#for example, in the above data set, male and female is catagorical data , 
#paletts will help to color with some specific color pallet 
```
