```python
#lmplots 
# they allow us to plot the linear model plots with seaborn 
# allow us to change aspect ratio 
import seaborn as sns
import matplotlib.pyplot as plt 
tips = sns.load_dataset("tips")
tips.head()
```

```python
#take x an y axis values 
sns.lmplot(x = 'total_bill', y = 'tip', data =tips , hue = 'sex', markers = ['o','v'],
          scatter_kws = {'s':100}) 
```

```python
# we can also use grid instead of hue
sns.lmplot(x = 'total_bill',y = 'tip',data = tips, col = 'sex')
# in the above code we have used grid = col and devided the dataset into 2 column by sex

```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips, col = 'sex', row = 'time')
# in the above code we have used grid = col and devided the dataset into 2 column by sex 
# and into 2 rows by time 
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips, col = 'day', hue = 'sex',aspect = 0.6,height = 8)
#the above code does not give perfect aspect ratio, so we have to do that by aspect = 0.6
```
