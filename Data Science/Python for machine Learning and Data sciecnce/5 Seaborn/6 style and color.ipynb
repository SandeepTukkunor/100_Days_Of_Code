```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_style("ticks")
sns.countplot(x = 'sex', data = tips)


```

```python
# we can also control the grid size using the matplotlib
plt.figure(figsize=(12,3))
sns.countplot(x = 'sex', data = tips)
```

```python
# controlling the font size 
#sns.set_context("poster") # this will zoom the context , but usual contect is notebook 
sns.set_context('notebook')

sns.countplot(x = 'sex', data = tips)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip',data = tips, hue = 'sex', palette = 'seismic')
# we can look for the pelette colors in google by searching 'matplotlib clolormap'
```
