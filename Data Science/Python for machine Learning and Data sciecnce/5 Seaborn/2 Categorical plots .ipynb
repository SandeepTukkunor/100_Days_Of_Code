```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
import numpy as np
```

```python
#most basic catogorical plot is "barplot"
sns.barplot(x= 'sex',y = "total_bill", data = tips,estimator=np.std)
#estimator helps 
```

```python
#countplot is similar to bar plot , but in this case we set only one value 
sns.countplot(x = 'sex', data = tips)
```

```python
#boxplot = shows the distribution of quantitative data 
sns.boxplot(x= 'day', y= 'total_bill', data = tips, hue = 'smoker')
```

```python
#it is almost similar to box plot 
sns.violinplot(x= 'day', y= 'total_bill', data = tips, hue = 'smoker',split = True)
```

```python
#stripplot is going to plot scatterig plot 
# x will be catagorical and y is numerical 
sns.stripplot(x = 'day', y = 'total_bill',data = tips, jitter = True, hue = "sex",dodge = True)
#jitter helps to know how many points are stacked on each other 
#parameter "split" has been renamed as dodge
```

```python
#swarm plot is similar to strip plot , difference is the points are adjusted so that they are not overlaped 
sns.swarmplot(x = 'day', y = 'total_bill',data = tips)
```

```python
# we can combine swarmplot with violinplot 
sns.violinplot(x = 'day', y = 'total_bill', data = tips)
sns.swarmplot(x = 'day', y = 'total_bill',data = tips, color = 'black')

```
