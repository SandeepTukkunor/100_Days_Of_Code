```python
# cleaning the data 
```

```python
plt.figure(figsize = (10,7))
sns.boxplot(x = "Pclass", y = "Age", data = train)
```

```python
sns.heatmap(data = train.isnull(), yticklabels=False, cbar=False, cmap= "viridis")
```

```python
# to fill the missing vause for Age 
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
      
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
        
    else:
        return Age
```

```python
train["Age"] = train[["Age", "Pclass"] ].apply(impute_age, axis = 1)
```

```python
sns.heatmap(data = train.isnull(), yticklabels=False, cbar=False, cmap= "viridis")
```

```python
# as a lot of data is missing, we can drop the cabin data 
train.drop("Cabin", inplace=True, axis = 1)
```

```python
sns.heatmap(data = train.isnull(), yticklabels=False, cbar=False, cmap= "viridis")
```

```python
train.dropna(inplace=True)
```

```python
sns.heatmap(data = train.isnull(), yticklabels=False, cbar=False, cmap= "viridis")
```

```python
# deal with categorical features 
```

```python
train.head()
```

```python
Sex = pd.get_dummies(train["Sex"], drop_first=True)
```

```python
# we have dropped first column becuse it will cause multi colinerity 
Sex.head()
```

```python
Embarked = pd.get_dummies(train["Embarked"], drop_first=True)
```

```python
Embarked.head()
```

```python
train = pd.concat([train, Sex, Embarked], axis = 1)
```

```python
train.head()
```

```python
#drop the columns which are not useful 
train.drop(["Sex", "Embarked", "Name", "Ticket"], axis = 1, inplace=True)
```

```python
train.head()
```

```python
#look if a data is missing
sns.heatmap(data = train.isnull(), yticklabels=False, cbar=False, cmap= "viridis")
```

```python
train.tail()
```

```python
train.drop(["PassengerId"], axis=1, inplace=True)
```

```python
# we are  done with cleaning the data, now we can train with machine learning algoriths
```

```python
train.head()
```

```python
X = train.drop("Survived", axis = 1)
y = train["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(predictions, y_test))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
```
