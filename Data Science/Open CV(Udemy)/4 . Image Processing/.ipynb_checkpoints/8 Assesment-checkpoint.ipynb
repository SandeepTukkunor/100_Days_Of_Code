```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def display_img(img, cmap = None):
    fig = plt.figure(figsize = (10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
#read img
img1 = cv2.imread("C:/Users/Sandeep Tukkunor/Desktop/Udemy Notes/Computer-Vision-with-Python/DATA/giraffes.jpg")
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
display_img(img1)
```

```python
#next task is to apply binary threshold 
# this is applied among all the three color channels
ret, thresh1 = cv2.threshold(img1 , 127, 255, cv2.THRESH_BINARY)
```

```python
display_img(thresh1)
```

```python
display_img(result)
```

```python

img = cv2.imread("C:/Users/Sandeep Tukkunor/Desktop/Udemy Notes/Computer-Vision-with-Python/DATA/giraffes.jpg")
```

```python
color = ['b', 'g', 'r']

for i, col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histr, color = col) 
    plt.xlim([0,256])
```
