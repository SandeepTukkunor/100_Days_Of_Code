```python
img = blue_bricks
color = ("b", "g", "r")

for i, col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histr, color = col) 
    plt.xlim([0,256])
    
    
    
plt.title("hstagram for blue bricks")
    


```

```python
rainbow= cv2.imread("C:/Users/Sandeep Tukkunor/Desktop/Udemy Notes/Computer-Vision-with-Python/DATA/rainbow.jpg")
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)
```

```python
img = rainbow
```

```python
img.shape
```

```python
#create a mask , using np
mask = np.zeros(img.shape[:2], np.uint8)
```

```python
#plt.imshow(mask, cmap = 'gray')
```

```python
mask[300:400, 100:400] = 255
```

```python
plt.imshow(mask,cmap ='gray')
```

```python
# bitwise operations
#create a masked image
masked_img = cv2.bitwise_and(img, img, mask = mask) #for histagram calculation
```

```python
plt.imshow(masked_img)  
```

```python
show_masked_img = cv2.bitwise_and(show_rainbow, show_rainbow, mask = mask) # visualization purpose
```

```python
plt.imshow(show_masked_img)
```

```python
hist_mask_values_red = cv2.calcHist([rainbow], channels = [2], mask = mask, histSize = [256], ranges = [0,256]) #channel = 2 , and in open cv its bgr red is at 2
```

```python
plt.plot(hist_mask_values_red)
```

```python
hist_values_red = cv2.calcHist([rainbow], channels = [2], mask = None, histSize = [256], ranges = [0,256])  #just for comparison we are removing mask
```

```python
plt.plot(hist_values_red)
```

```python
gorilla = cv2.imread("C:/Users/Sandeep Tukkunor/Desktop/Udemy Notes/Computer-Vision-with-Python/DATA/gorilla.jpg", 0)
```

```python
plt.imshow(gorilla)
```

```python
def display(img , cmap = None):
    fig = plt.figure(figsize = (10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
    
```

```python
display(gorilla, cmap = 'gray')
```

```python
gorilla.shape
```

```python
hist_values = cv2.calcHist([gorilla],channels =  [0],mask = None, histSize = [256], ranges = [0,256])
```

```python
plt.plot(hist_values)
```

```python
#equalize the hist 
#equalize basically means, increasing the contrast 
eq_gorilla = cv2.equalizeHist(gorilla)
```

```python
display(eq_gorilla, cmap = 'gray')
```

```python
hist_values1 = cv2.calcHist([eq_gorilla],channels =  [0],mask = None, histSize = [256], ranges = [0,256])
```

```python
plt.plot(hist_values1)
```
