```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Mall_Customers.csv")
```

```python
data.head()
```

```python
X = data.iloc[:, [3,4]].values
```

```python
X
```

```python
#using the dedrogram to find the optimal number of clusters
```

```python
#this is not imported from sci-kit learn but imported from sci-py
```

```python

plt.figure(figsize=(35,20))
import scipy.cluster.hierarchy  as sch
dendrogram = sch.dendrogram(sch.linkage(X, method="ward"))
plt.xlabel("Customers")
plt.ylabel("Euclidian distance ")
```

```python

#we have identified that , the best number of clusters is 5
#build, train and run hierarchial clustering 
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity="euclidean", linkage="ward")
y_sc = hc.fit_predict(X)
```

```python
y_sc
```

```python
X[y_sc ==0, 0]
```

```python
X[y_sc ==0, 1]
```

```python
plt.scatter(X[y_sc ==0, 0], X[y_sc == 0 , 1], c = 'red', label = "cluster 1")
plt.scatter(X[y_sc ==1, 0], X[y_sc == 1 , 1], c = 'green', label = "cluster 1")
plt.scatter(X[y_sc ==2, 0], X[y_sc == 2 , 1], c = 'blue', label = "cluster 1")
plt.scatter(X[y_sc ==3, 0], X[y_sc == 3 , 1], c = 'black', label = "cluster 1")
plt.scatter(X[y_sc ==4, 0], X[y_sc == 4 , 1], c = 'brown', label = "cluster 1")
plt.legend()
```
